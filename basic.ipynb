```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def sinplot():
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.cos(x+1*0.5)*(7- i))
        plt.plot(x, np.sin(x+1)*(7- i))
```

```python
sinplot()
```
